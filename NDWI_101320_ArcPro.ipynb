### Extracting the Bands 

NDWI is done with the equation above. The NAIP has 4 bands. Red, Green, Blue and Near infra Red. This are diffrenet wavelengther of light that the sensor collects based on the surfaces that they reflect off. To find water green and NIR are used because green is flected by the water and NIR is abosorbed and make for very distict profile.
